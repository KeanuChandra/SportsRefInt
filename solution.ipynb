```python
import pandas as pd
import ast

#read json as dict
with open('data.json', 'r') as f:
    file_content = f.read()

data_dict = ast.literal_eval(file_content)
```

```python
#convert dict to dataframe
df = pd.DataFrame.from_dict(data_dict, orient='index')
#Replace cell values with their corresponding 'W' value. If value is NaN leave it as is.
df = df.map(lambda x: x.get('W') if isinstance(x, dict) else x)
#convert to int64 to remove trailing decimal
df = df.astype('Int64')
```

```python
#arrange in requested order
new_order = ['BRO', 'BSN', 'CHC', 'CIN', 'NYG', 'PHI', 'PIT', 'STL']
df = df.loc[new_order]
df = df[new_order]
#show table
print(df)
```
